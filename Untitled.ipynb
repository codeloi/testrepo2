```python
import networkx as nx
import matplotlib.pyplot as plt

# Tạo đồ thị và thêm cạnh
G = nx.Graph()
edges = [(1, 2), (1, 3), (1, 4), (2, 3), (2, 4), (2, 5), (3, 4), (3, 5), (4, 5), (5, 6)]
G.add_edges_from(edges)

# Vẽ đồ thị
pos = nx.spring_layout(G)  # Tạo bố cục cho đồ thị
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='grey')
plt.title('Đồ thị đơn vô hướng với DFS', fontsize=14)

# Mô tả quá trình chạy DFS
def DFS(G, start, visited=None):
    if visited is None:
        visited = set()
    visited.add(start)
    print(start, end=' ')

    for next in set(G[start]) - visited:
        DFS(G, next, visited)
    return visited

print('Quá trình chạy DFS:')
DFS(G, 1)
plt.show()




```
